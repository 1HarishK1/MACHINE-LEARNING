##ACCURACY AND VALIDATION OF HEART FAILURE CLINICAL RECORD##

#Importing Libraries

```python
import pandas as pd
```

```python
dataset=pd.read_csv("/content/drive/MyDrive/heart_failure_clinical_records_dataset.csv")
```

#Loading dataset

```python
dataset.head()
```

#Finding the Matrix strucuture of Imported Dataset

```python
print(dataset.shape)
```

#Creating Values of X and Y Variable

```python
x=dataset.iloc[:,0:12].values
print(x)
y=dataset.iloc[:,-1].values
print(y)
```

```python
y
```

#Splitting the Dataset for Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

#Feature Scaling By StandardScaler

```python
from sklearn.preprocessing import StandardScaler as sc
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(x_train)
```

#Training the imported dataset By Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

#*Predicting For All  Input Test Datas*

```python
y_pred=model.predict(x_test)
y_pred
```

#The Accuracy of the Imported Dataset

```python
from sklearn.metrics import  accuracy_score
print("Accuracy of the Model is : ","{:.2f}".format(accuracy_score(y_test,y_pred)*100))
```
